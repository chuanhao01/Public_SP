```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading in the dataset
resale_prices = pd.read_csv('data/resale-flat-prices-based-on-approval-date-1990-onwards.csv', parse_dates=['month'])
```

```python
# Setting the context for the whole notebook
sns.set_context(rc={"font.size":18, "axes.titlesize":18, "axes.labelsize":18, 'xtick.labelsize': 15, 'ytick.labelsize': 15})
sns.set()
```

# Looking at the data on their own

```python
display('Resale prices head')
display(resale_prices.head())
display('Basic stats of the resale prices')
display(resale_prices.describe(include='all'))
display('Checking if there is any missing/null values')
display(resale_prices.isnull().sum())
```

# Reshaping the data and cleaning

```python
# Creating map to map town by regions
def splitRegion(region_str, region):
    arr = region_str.upper()
    arr = arr.split(', ')
    arr = {i:region for i in arr}
    return arr

north = 'Admirality, Kranji, Woodlands, Sembawang, Yishun, Yio Chu Kang, Seletar, Sengkang'
north = splitRegion(north, 'north')

south = 'Holland, Queenstown, Bukit Merah, Telok Blangah, Pasir Panjang, Sentosa, Bukit Timah, Newton, Orchard, City, Marina South'
south = splitRegion(south, 'south')

east = 'Serangoon, Punggol, Hougang, Tampines, Pasir Ris, Loyang, Simei, Kallang, Katong, East Coast, Macpherson, Bedok, Pulau Ubin, Pulau Tekong'
east = splitRegion(east, 'east')

west = 'Lim Chu Kang, Choa Chu Kang, Bukit Panjang, Tuas, Jurong East, Jurong West, Jurong Industrial Estate, Bukit Batok, Hillview, West Coast, Clementi'
west = splitRegion(west, 'west')

central = 'Thomson, Marymount, Sin Ming, Ang Mo Kio, Bishan, Serangoon Gardens, MacRitchie, Toa Payoh'
central = splitRegion(central, 'central')

all_regions = {**north, **south, **east, **west, **central}
```

```python
# Here we are reshaping the resale data to be by year
# Replace spelling errors
resale_prices = resale_prices.replace({'MULTI-GENERATION': 'MULTI GENERATION'})
# Make the resale prices by year to smooth out values
df_resale = resale_prices.groupby([pd.Grouper(freq='Y', key='month'), 'town', 'flat_type', 'storey_range', 'lease_commence_date'])['resale_price', 'floor_area_sqm'].mean().reset_index()

# Creating the regions column
df_resale['regions'] = df_resale['town'].map(all_regions)
# 'Creating the column year left on lease
df_resale['lease_years_left'] = 99 - 2020 + df_resale['lease_commence_date']

# Creating price per sqm
df_resale['price_per_sqm'] = df_resale['resale_price'] / df_resale['floor_area_sqm']
display(df_resale.head())
```

# Starting the anaylsis

```python
# Looking at the resale price by regions and town

# Getting data by regions
by_regions = df_resale.groupby(['month', 'regions'])['resale_price'].mean().reset_index()

# Creating the subplots
fig, ax = plt.subplots(2, 2, figsize=(20, 20))

# Getting the first 2 subplots by regions
sns.violinplot(x='regions', y='resale_price', data=by_regions, ax=ax[0][0])
sns.lineplot(x='month', y='resale_price', data=by_regions, hue='regions', ax=ax[0][1])
plt.setp(ax[0][1].get_legend().get_texts(), fontsize='15')

# Setting up the titles for first 2 
ax[0][0].title.set_text('Violinplot of the average resale prices of all flats by region')
ax[0][1].title.set_text('Line plot of the average resale prices by region over time')

# Filtering to only have the south regions
only_south = df_resale[df_resale['regions'] == 'south'].groupby(['month', 'town'])['resale_price'].mean().reset_index()

# Getting the last 2 subplots by the south region
sns.violinplot(x='town', y='resale_price', data=only_south, ax=ax[1][0])
sns.lineplot(x='month', y='resale_price', data=only_south, hue='town', ax=ax[1][1])
plt.setp(ax[1][1].get_legend().get_texts(), fontsize='15')

# Setting up titles for next 2
ax[1][0].title.set_text('Violinplot of the average resale prices of all flats\nin the south region by town')
ax[1][1].title.set_text('Line plot of the average resale prices by town over time')

# Saving the figure
plt.savefig('final_plots/resale_1.png', bbox_inches='tight')
pass
```

From figure 2 we can see that the trends of the average resale prices for all regions tend to have the same general trend.
However, we can also see over time that the ‘south’ region has the highest resale price compared to the other regions.

Comparing that to figure 1, we see that the violinplot for ‘south’ is also the longest, showing that it has the largest distribution and range of prices.
We can also see that the rest tend to not vary as much and have long bottom tails which is likely due to low prices at the beginning.

This tells us that the demand for resale flats in the 'south' region is liekly to be higer compared to the reset of the regions, as seen from the average prices being higher than the reset since 2008.
Since all the regions would have the same trend , any increase of decrease in resale prices would also somewhat affect all regions, making South still have the highest resale price in the future.
Taking into account % growth, the value of a flat in the South would increase more than the rest and would be valued higher than the rest. 

Thus we would pick the South region to buy a flat in.

From figure 4, we can see that similar to the region example, the average resale price of flats in Bukit Timah is much higher compared to the other 2 towns in the 'South' region.
Further more, it seems that the resale prices in Bukit Timah is increasing currently as well.
We can also see that the towns in the 'South' region also tend to have quite a high price, which would also allow the resale prices of flats in Buki Timah to remain high or increase further.
Looking at figure 3, we see the same trend as before, with Bukit Timah having the longest violinplot, which is most likely due to it average resale price beinh higher than the rest.

As the flats in Bukit Timah have such a higher price compared to the other towns in the 'South' region, we can see that the demand for flats in Bukit Timah is the highest. And since the resale price of other towns are not far behind, we can see that the demand will remain high and could even increase.
Making buying a flat in Bukit Timah the best option for towns.
Therefore, we would pick to buy a flat in Bukit Timah.

```python
# Looking at difference of resale prices by flat type in Bukit Timah

# Filtering data to get only resale flats in the bukit timah
# Then doing some data mainipulation to get data to plot
confirm_town = df_resale[df_resale['town'] == 'BUKIT TIMAH']
# To see resale price by flat type
df_flat_type = confirm_town.groupby(['month', 'flat_type'])['resale_price', 'price_per_sqm'].mean().reset_index()
# To see the change in the resale price as a % change
pct_change_flat = df_flat_type.pivot(index='month', columns='flat_type')['resale_price'].pct_change(periods=3).reset_index().melt(id_vars='month')

# Getting the subplots
fig, ax = plt.subplots(2, 2, figsize=(20, 20))
# The first 2 plots
sns.lineplot(x='month', y='resale_price', data=df_flat_type, hue='flat_type', ax=ax[0][0])
sns.lineplot(x='month', y='value', data=pct_change_flat , hue='flat_type', ax=ax[0][1])
plt.setp(ax[0][0].get_legend().get_texts(), fontsize='15')
plt.setp(ax[0][1].get_legend().get_texts(), fontsize='15')

# Adding titles
ax[0][0].title.set_text('Line plot of the resale prices of flats\nin Bukit Timah by flat type')
ax[0][1].title.set_text('Line plot of the change in resale prices of flats\nin Bukit Timah by flat type as a ratio')

sns.lineplot(x='month', y='price_per_sqm', data=df_flat_type, hue='flat_type', ax=ax[1][0])
sns.violinplot(x='flat_type', y='resale_price', data=df_flat_type, ax=ax[1][1])

# Adding titles
ax[1][0].title.set_text('Line plot of the price per sqm of flats\nin Bukit Timah by flat type')
ax[1][1].title.set_text('Violinplot of the resale prices of flats\nin Bukit Timah by flat types')

# Saving the figure
plt.savefig('final_plots/resale_2.png', bbox_inches='tight')
pass

# Same thing here, we can see that the flat types seem to all have the same general trend
# Note executive sharp increase, after 5-room sharp increase
# Look at %change in price. Other than 3, the rest seem to be about the same
# Middle where 5-room and executive difference?
# Looking at worth, we can 4-room and 5-room are most worth

# 3-room fluctuates the most
```

- Figure 1
    - All flat types also seem to follow the same trend
    - Difference in prices is due to how large the flat is
    - However there is a change in recent years
    - Note how 5 rooms were almost as expensive as executive flats
    - The gap of 4 and 5 room flats decreased as well.
- Figure 2
    - We can see that the lines do somewhat follow the same trend until recently
    - Executive flats resale prices changes are also more volatile
- Figure 3
    - This is more looking at the worth of a flat
    - We see that now, the worth of a 4, 5 room and executive flat are around the same.
    - But a 3 room flat is much cheaper, but the prices are increasing
- Figure 4
    - This mainly reinforces figure 2 and 3, where the resale prices of 4, 5 and executive flats have the same shape.
- Conclusions:
    - 4, 5 and executive flats are worth around the same
    - 3 Room flats are cheaper but worth lesser.
    - 3 room flats also tend not to vary, not much as an investment
    - Not considering executive, as it is mainly too expensive upfront
    - Thus we are looking to buy a 4 or 5 Room flat in Bukit Timah


```python
# Looking at the difference of resale prices by number of lease years remaining

# Getting the boolean mask for 4 and 5 room
room_4_filter = confirm_town['flat_type'] == '4 ROOM'
room_5_filter = confirm_town['flat_type'] == '5 ROOM'
flat_type_filter = room_4_filter | room_5_filter
# Filtering based on only 4 and 5 room
confirm_town_flat = confirm_town[flat_type_filter]

# Getting the dataset to be by lease years for 4 and 5 rooms in bukit timah
confirm_lease = confirm_town_flat.groupby(['month', 'flat_type', 'lease_years_left'])['resale_price'].mean().reset_index()
# Same as above but for all and all flat types
all_lease = df_resale.groupby(['month', 'lease_years_left'])['resale_price'].mean().reset_index()

# Setting up the subplots
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Plotting the scatter plots
sns.scatterplot(x='month', y='resale_price', size='lease_years_left', hue='flat_type', data=confirm_lease, ax=ax[0])
sns.scatterplot(x='month', y='resale_price', size='lease_years_left', data=all_lease, ax=ax[1])

# Setting up the axis
plt.setp(ax[0].get_legend().get_texts(), fontsize='15')
plt.setp(ax[1].get_legend().get_texts(), fontsize='15')

# Adding titles
ax[0].title.set_text('Scatter plot of 4 and 5 Room flats in Bukit Timah\nof resale prices over time and number of lease years remaining')
ax[1].title.set_text('Scatter plot of resale prices of flats over time\nand number of lease years remaining')

# Saving the figure
plt.savefig('final_plots/resale_3_1.png', bbox_inches='tight')
pass
```

- Figure 1
    - We can see that for 5 Room flats, a flat with more lease years remaining is cheaper than a flat with lesser lease years remaining
    - The reverse is true for 4 Room flats where flats with more lease years are more expensive
    - Gap in high number of lease years left of 4 and 5 Room is not that much
- Figure 2
    - Generally, flats with a longer number of lease years left are more expensive. (Larger dots are higher)
- Figure 3
    - With a KDE plot of number of lease years left with resale price, we can see that flats with a smaller number of lease years left are generally cheaper.
    - Seen from shaded region tending towards a downwards left.
- Figure 4
    - KDE for 4 and 5 Rooms flats in Bukit Timah shows a shaded region being a lower right for the 5 Room flats, showing higher number of lease years are cheaper for 5 Room flats
- Conclusions:
    - Cheaper to buy a younger 5 room flat.
    - Compared to trends generally, this flat would be worth more, even if it is not worth as much in Bukit Timah.
    - The old 5 room flat that is more expensive may be an outlier.
    - Thus we would want to buy a younger 5 Room Flat in Bukit Timah.


```python
# For the kde comparing 4 and 5 room
# Getting boolean mask for filtering
room_4_filter = confirm_lease['flat_type'] == '4 ROOM'
room_5_filter = confirm_lease['flat_type'] == '5 ROOM'

# Filter the lease data by r4 and 5 room
only_4_room = confirm_lease[room_4_filter]
only_5_room = confirm_lease[room_5_filter]

# Getting subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# Plotting kdes
sns.kdeplot(only_4_room['lease_years_left'], only_4_room['resale_price'], ax=ax[0], cmap='Reds', shade=True, shade_lowest=False)
sns.kdeplot(only_5_room['lease_years_left'], only_5_room['resale_price'], ax=ax[1], cmap='Blues', shade=True, shade_lowest=False)

# Setting titles and removing the x-axis label
ax[0].title.set_text('KDE plot of resale prices to number of lease years left\nof 4 Rooms flats in Bukit Timah')
ax[1].title.set_text('KDE plot of resale prices to number of lease years left\nof 5 Rooms flats in Bukit Timah')

for a in ax.flat:
    a.label_outer()
    
# Saving the figure
plt.savefig('final_plots/resale_3_2.png', bbox_inches='tight')
    
# For all flat types
# Getting the subplots
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plotting the data
sns.kdeplot(all_lease['lease_years_left'], all_lease['resale_price'], ax=ax, cmap='Reds', shade=True, shade_lowest=False)

# Setting the title
ax.title.set_text('KDE plot of resale prices to number of lease years left\n for all flats')

# Saving the figure
plt.savefig('final_plots/resale_3_3.png', bbox_inches='tight')
```

```python
# Looking difference of resale price by storey 
# Filtering all flats for only 5 room
all_5_room_filter = df_resale['flat_type'] == '5 ROOM'
all_5_room = df_resale[all_5_room_filter]
all_5_room_storey = all_5_room.groupby(['month', 'storey_range'])['resale_price'].mean().reset_index()
all_5_room_storey = all_5_room_storey.sort_values(by='storey_range')

# Filtering data for only 5 rooms in bukit timah
confirm_5_room_filter = confirm_town_flat['flat_type'] == '5 ROOM'
confirm_5_room = confirm_town_flat[confirm_5_room_filter]
confirm_5_room_storey = confirm_5_room.groupby(['month', 'storey_range'])['resale_price'].mean().reset_index()
confirm_5_room_storey = confirm_5_room_storey.sort_values(by='storey_range')

# Setting up the subplots
fig, ax = plt.subplots(2, 1, figsize=(20, 20))

# Plotting the lineplots
sns.lineplot('month', 'resale_price', data=all_5_room_storey, hue='storey_range', ax=ax[0], palette=sns.color_palette("Reds", n_colors=23))
sns.lineplot('month', 'resale_price', data=confirm_5_room_storey, hue='storey_range', ax=ax[1], palette=sns.color_palette("Reds", n_colors=11))

# Setting titles and removing the x-axis label
ax[0].title.set_text('Line plot of 5 Room flats resale prices over time by storey ranges')
ax[1].title.set_text('Line plot of 5 Room flats in Bukit Timah resale prices over time by storey ranges')

# Saving the figure
plt.savefig('final_plots/resale_4.png', bbox_inches='tight')
```

- Figure 1
    - We can see a very clear trend that the higher the flat storey is, the more expensive it tends to be across Singapore.
    - However we can see that there are ‘groups’ of storeys that clump with each other.
- Figure 2
    - This trend is also reflected in 5 Room flats in Bukit Timah as well
    - However besides the outliers of the higher storey and lowest storey, the other storey seem to have generally the same prices
    - Also we see that the prices tend to really stick to the same trend, thus we probably would not expect much in a flats price to come from the storey.
- Conclusions:
    - The storey of the 5 Room flat would both come down to preference and how much more the family would be willing to pay.
    - It might be a better investment to invest in the higher floors, but the data in Bukit Timah suggest they are around the same price, which could be that they are in the same storey ‘group’ 
    - Thus we would pick 10-12 storey for us.
