```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels.formula.api import ols
```

```python
prop_info = pd.read_csv('own_data/hdb-property-information.csv', parse_dates=['year_completed'])
area_dwell = pd.read_csv('own_data/land-area-and-dwelling-units-by-town.csv', parse_dates=['financial_year'])
```

# Format

```python
# Converting codes to actual names
town_format = {'AMK': 'ANG MO KIO',
 'BB': 'BUKIT BATOK',
 'BD': 'BEDOK',
 'BH': 'BISHAN',
 'BM': 'BUKIT MERAH',
 'BP': 'BUKIT PANJANG',
 'BT': 'BUKIT TIMAH',
 'CCK': 'CHOA CHU KANG',
 'CL': 'CLEMENTI',
 'CT': 'CENTRAL AREA',
 'GL': 'GEYLANG',
 'HG': 'HOUGANG',
 'JE': 'JURONG EAST',
 'JW': 'JURONG WEST',
 'KWN': 'KALLANG/WHAMPOA',
 'MP': 'MARINE PARADE',
 'PG': 'PUNGGOL',
 'PRC': 'PASIR RIS',
 'QT': 'QUEENSTOWN',
 'SB': 'SEMBAWANG',
 'SGN': 'SERANGOON',
 'SK': 'SENGKANG',
 'TAP': 'TAMPINES',
 'TG': 'TENGAH',
 'TP': 'TOA PAYOH',
 'WL': 'WOODLANDS',
 'YS': 'YISHUN'}

prop_info = prop_info.replace(town_format)
```

```python
# Creating map to map town by regions
def splitRegion(region_str, region):
    arr = region_str.upper()
    arr = arr.split(', ')
    arr = {i:region for i in arr}
    return arr

north = 'Admirality, Kranji, Woodlands, Sembawang, Yishun, Yio Chu Kang, Seletar, Sengkang, NORTH'
north = splitRegion(north, 'north')

south = 'Holland, Queenstown, Bukit Merah, Telok Blangah, Pasir Panjang, Sentosa, Bukit Timah, Newton, Orchard, City, Marina South, SOUTH'
south = splitRegion(south, 'south')

east = 'Serangoon, Punggol, Hougang, Tampines, Pasir Ris, Loyang, Simei, Kallang, Katong, East Coast, Macpherson, Bedok, Pulau Ubin, Pulau Tekong, KALLANG/WHAMPOA, MARINE PARADE, KALLANG/ WHAMPOA, EAST'
east = splitRegion(east, 'east')

west = 'Lim Chu Kang, Choa Chu Kang, Bukit Panjang, Tuas, Jurong East, Jurong West, Jurong Industrial Estate, Bukit Batok, Hillview, West Coast, Clementi, TENGAH, WEST'
west = splitRegion(west, 'west')

central = 'Thomson, Marymount, Sin Ming, Ang Mo Kio, Bishan, Serangoon Gardens, MacRitchie, Toa Payoh, CENTRAL AREA, GEYLANG, OTHER ESTATES, CENTRAL'
central = splitRegion(central, 'central')

all_regions = {**north, **south, **east, **west, **central}
```

```python
prop_info['regions'] = prop_info['bldg_contract_town'].replace(all_regions)
prop_info['regions'].unique()
```

```python
y_n_format = {'N': 0, 'Y': 1}
prop_info = prop_info.replace(y_n_format)
```

# Area dwell

```python
area_dwell['town'] = area_dwell['town'].str.upper()
area_dwell = area_dwell.replace(all_regions)
area_dwell = area_dwell.replace({'-': 0})
area_dwell = area_dwell.replace({'na': 0})
area_dwell = area_dwell.astype({'dwelling_units_under_management': 'int32'})
area_dwell = area_dwell.astype({'total_land_area': 'int32'})
```

# Anaylsis below

```python
regions_amn = prop_info.groupby(['year_completed', 'regions'])['residential','commercial', 'market_hawker', 'miscellaneous', 'multistorey_carpark','precinct_pavilion'].sum().reset_index()
regions_amn
```

```python
regions_amn['amenities'] = regions_amn[['commercial', 'market_hawker', 'miscellaneous', 'multistorey_carpark','precinct_pavilion']].sum(axis=1)
regions_amn['amenities_sum'] = regions_amn.groupby(['year_completed', 'regions'])['amenities'].sum().groupby(['regions']).cumsum().reset_index()['amenities']
regions_amn['residential_sum'] = regions_amn.groupby(['year_completed', 'regions'])['residential'].sum().groupby(['regions']).cumsum().reset_index()['residential']
regions_amn['am_re_ratio'] = regions_amn['amenities_sum'] / regions_amn['residential_sum']
regions_amn
```

```python
fig, ax = plt.subplots(3, 1, figsize=(20, 20))
sns.lineplot(x='year_completed', y='amenities_sum', hue='regions', data=regions_amn, ax=ax[0])
sns.lineplot(x='year_completed', y='residential_sum', hue='regions', data=regions_amn, ax=ax[1])
sns.lineplot(x='year_completed', y='am_re_ratio', hue='regions', data=regions_amn, ax=ax[2])
```

```python
# sns.heatmap(regions_amn[['year_completed', 'regions', 'amenities']])
a = regions_amn[['year_completed', 'regions', 'amenities']]
a = a.groupby([pd.Grouper(freq='5Y', key='year_completed'), 'regions'])['amenities'].sum().reset_index()
# a = a.set_index('year_completed')
a['year_completed'] = a['year_completed'].dt.strftime('%Y')
a = a.pivot('year_completed', 'regions', 'amenities')
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
sns.heatmap(a, annot=True, fmt='.1f', ax=ax)
```

```python
a = regions_amn[['year_completed', 'regions', 'amenities', 'amenities_sum', 'residential_sum', 'am_re_ratio']]
b = a[['year_completed', 'regions', 'amenities_sum']]
c = b.copy()
b['year_completed'] = b['year_completed'].dt.strftime('%Y')
# b.pivot('year_completed', 'regions', 'amenities_sum').plot.bar(stacked=True, figsize=(10, 10))

new_index = pd.MultiIndex.from_product([a['year_completed'].unique(), a['regions'].unique()], names=['year_completed', 'regions'])
# d = a.groupby(['year_completed', 'regions'])['amenities_sum'].sum().reindex(new_index).reset_index().groupby(['regions', 'year_completed'])['amenities_sum'].apply(lambda x: x.fillna(method='ffill'))
d = a.groupby(['year_completed', 'regions'])['amenities_sum'].sum().reindex(new_index).reset_index()
d['amenities_sum'] = d.groupby(['regions'])['amenities_sum'].fillna(method='ffill')
d = d.fillna(0)
d['year_completed'] = d['year_completed'].dt.strftime('%Y')
d.pivot('year_completed', 'regions', 'amenities_sum').plot.bar(stacked=True, figsize=(20, 10))

e = a.groupby(['year_completed', 'regions'])['am_re_ratio'].mean().reindex(new_index).reset_index()
e['am_re_ratio'] = e.groupby(['regions'])['am_re_ratio'].fillna(method='ffill')
e = e.fillna(0)
e = e.groupby([pd.Grouper(freq='10Y', key='year_completed'), 'regions'])['am_re_ratio'].mean().reset_index()
e['year_completed'] = e['year_completed'].dt.strftime('%Y')
e = e.pivot('year_completed', 'regions', 'am_re_ratio')
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
sns.heatmap(e, ax=ax, annot=True, fmt='.2f')
```
