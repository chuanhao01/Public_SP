```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# sns.set(font_scale=2)
sns.set()
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].text(0.25, 0.45, 'Figure 1', fontsize=40)
ax[0, 1].text(0.25, 0.45, 'Figure 2', fontsize=40)
ax[1, 0].text(0.25, 0.45, 'Figure 3', fontsize=40)
ax[1, 1].text(0.25, 0.45, 'Figure 4', fontsize=40)

plt.savefig('final_plots/figure_demo_1.png', bbox_inches='tight')
```
